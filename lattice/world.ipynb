```python
import matplotlib.pyplot as plt
import seaborn as sns
from world import World  # pyrefly: ignore[missing-import]

world = World(33, 30)
world.fill_rectangle(
    8, 15, 11, 4, 1
)  # Fill a rectangle from (0, 0) to (6, 6) with block type 1


world.fill_rectangle(
    7, 10, 6, 9, 2
)

data = world.grid

sns.heatmap(
    data,
    cmap="binary",
    cbar=False,
    linewidths=2,  # linewidths añade el espacio (rejilla) entre cuadros
    linecolor="black",
    square=True,
    xticklabels=range(1, world.width + 1),  # type: ignore
    yticklabels=range(1, world.height + 1),  # type: ignore
)

plt.show()
```
