```python
import matplotlib.pyplot as plt
import seaborn as sns
from world import World  # pyrefly: ignore[missing-import]

world = World(10, 10)
world.fill_reactangle(
    0, 0, 6, 6, 1
)  # Fill a rectangle from (0, 0) to (6, 6) with block type 1

data = world.grid

sns.heatmap(
    data,
    cmap="binary",
    cbar=False,
    linewidths=2,  # linewidths añade el espacio (rejilla) entre cuadros
    linecolor="black",
    square=True,
)

plt.show()
```
