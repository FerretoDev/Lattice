```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Parámetros
g = 9.81      # gravedad
L = 1.0       # longitud del péndulo (m)
dt = 0.05     # paso de tiempo

# Condiciones iniciales
theta = np.pi / 4   # ángulo inicial (45°)
omega = 0.0         # velocidad angular

# Listas para guardar datos
thetas = []

# Simulación (método de Euler)
for _ in range(300):
    alpha = -(g / L) * np.sin(theta)  # aceleración angular
    omega += alpha * dt
    theta += omega * dt
    thetas.append(theta)

# Convertir a coordenadas cartesianas
x = L * np.sin(thetas)
y = -L * np.cos(thetas)

# Crear figura
fig, ax = plt.subplots()
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 0.2)

line, = ax.plot([], [], 'o-', lw=2)

def update(frame):
    line.set_data([0, x[frame]], [0, y[frame]])
    return line,

ani = FuncAnimation(fig, update, frames=len(x), interval=50)

plt.title("Simulación de un péndulo")
plt.show()
```
