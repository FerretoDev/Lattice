```python
import matplotlib.pyplot as plt
import seaborn as sns
from world import World  # pyrefly: ignore[missing-import]

world = World(50, 50)
world.fill_rectangle(
    5, 5, 8, 9, 1
)  # Fill a rectangle from (0, 0) to (6, 6) with block type 1
print(world.counter_blocks)

data = world.grid

sns.heatmap(
    data,
    cmap="cividis",  # Change the color map to 'binary' for a black and white representation
    cbar=False,
    linewidths=2,  # linewidths añade el espacio (rejilla) entre cuadros
    linecolor="black",
    square=True,
    xticklabels=range(1, world.width + 1),  # type: ignore
    yticklabels=range(1, world.height + 1),  # type: ignore
)

plt.show()
```

```python
world = World(10, 10)
world.set_block(2, 3, 1)  # Setting block type 1 at (2, 3)

data = world.grid

fig, ax = plt.subplots()

sns.heatmap(
    data,
    cmap="binary",
    cbar=False,
    linewidths=0,  # Sin líneas automáticas, las dibujamos manualmente
    square=True,
    xticklabels=range(1, world.width + 1),  # type: ignore
    yticklabels=range(1, world.height + 1),  # type: ignore
    ax=ax,
)

# Dibuja líneas de grilla con color según el valor de la celda:
# negro (1) → líneas blancas, blanco (0) → líneas negras
for y in range(world.height):
    for x in range(world.width):
        line_color = "white" if data[y, x] == 1 else "black"
        ax.add_patch(plt.Rectangle((x, y), 1, 1, fill=False, edgecolor=line_color, linewidth=2))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Creamos una matriz de ejemplo (4x4)
# 1 = Negro, 0 = Blanco
def crear_matriz_binaria(ancho, alto):
    """Crea una matriz binaria (0 y 1) con el tamaño indicado."""
    return np.random.randint(0, 2, size=(alto, ancho)).tolist()

# Solo cambias estos dos valores:
data = crear_matriz_binaria(ancho=4, alto=4)

# 2. Generamos el gráfico
# 'gray_r' invierte la escala de grises: 0 es blanco y 1 es negro
plt.imshow(data, cmap="gray_r")

# 3. Estética opcional
plt.title("Matriz Binaria (1=Negro, 0=Blanco)")
plt.axis("off")  # Oculta los números de los ejes si no los necesitas

plt.show()
```
